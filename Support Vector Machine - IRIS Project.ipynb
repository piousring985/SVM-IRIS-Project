### Import of various required libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import datasets
from sklearn import svm
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('whitegrid')
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
col = ['petal_length', 'petal_width', 'species']
df.loc[:, col].head()
```

```python
col = ['petal_length', 'petal_width']
X = df.loc[:, col]
```

```python
species_to_num = {'setosa':0,
                 'versicolor': 1,
                 'virginica':2}
df['tmp'] = df['species'].map(species_to_num)
y = df['tmp']
```

```python
X.shape
```

### Application of SVM - Kernel = Linear

```python
C = 1
clf =svm.SVC(kernel='linear', C=C)
clf.fit(X, y)
```

```python
clf.predict([[6, 2]])
```

### Visualization of Support Vector Machine

```python
Xv = X.values.reshape(-1,1)
h = 0.02
x_min, x_max = Xv.min(), Xv.max() + 1
y_min, y_max = y.min(), y.max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
```

```python
z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)
fig = plt.figure(figsize=(16,10))
ax = plt.contourf(xx, yy, z, cmap = 'afmhot', alpha=0.3);
plt.scatter(X.values[:, 0], X.values[:, 1], c=y, s=80, 
            alpha=0.9, edgecolors='g');
```

```python
df = sns.load_dataset('iris')
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
col = ['petal_length', 'petal_width']
X = df.loc[:, col]
species_to_num ={'setosa':0, 
                'versicolor':1,
                'virginica':2}
df['tmp'] = df['species'].map(species_to_num)
y = df['tmp']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=0)
```

```python
sc_x = StandardScaler()
X_std_train = sc_x.fit_transform(X_train)
```

```python
C = 1.0
clf = svm.SVC(kernel ='linear', C=C)
clf.fit(X_std_train, y_train)
```

#### Model Evaluation

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
res = cross_val_score(clf, X_std_train, y_train, cv=10, scoring = 'accuracy')
print('Agerage Accuracy: \t {0:.4f}'.format(np.mean(res)))
print('Accuracy SD: \t\t {0:.4f}'.format(np.std(res)))
```

```python
y_train_pred = cross_val_predict(clf, X_std_train, y_train, cv=3)
```

```python
confusion_matrix(y_train, y_train_pred)
```

```python
print("Precision Score: \t {0:.4f}".format(precision_score(y_train, 
                                                           y_train_pred, 
                                                           average='weighted')))
print("Recall Score: \t\t {0:.4f}".format(recall_score(y_train,
                                                     y_train_pred, 
                                                     average='weighted')))
print("F1 Score: \t\t {0:.4f}".format(f1_score(y_train,
                                             y_train_pred, 
                                             average='weighted')))
```

```python
y_test_pred = cross_val_predict(clf, sc_x.transform(X_test), y_test, cv=3)
```

```python
print('precisoin Score: \t {0:.4f}'.format(precision_score(y_test,
                                                          y_test_pred,
                                                          average ='weighted')))
print('Recall Score: \t\t {0:.4f}'.format(recall_score(y_test, y_test_pred,
                                                      average='weighted')))
print('F1 score: \t\t {0:.4f}'.format(f1_score(y_test, y_test_pred, 
                                              average='weighted')))
```
